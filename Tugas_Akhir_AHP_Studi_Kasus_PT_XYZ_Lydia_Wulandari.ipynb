## Step 3 : Membuat Matriks Perbandingan Berpasangan Terhadap Kriteria (Pairwise Comparison Matrix)

```python
#STEP 3.1 Perbandingan berpasangan setiap kriteria

#Untuk setiap kriteria dilakukan perbandingan dengan kriteria lainnya, akan didapatkan pasangan:
#(JP,JAM), (JP,HPA), (JP,WP), (JP,UP), (JAM,HPA), (JAM,WP), (JAM,UP), (HPA,WP), (HPA,UP), (WP,UP)
#Seluruh pasangan disimpan di dalam sebuah tabel matriks.
```

```python
#STEP 3.2 Penilaian Expert Terhadap Kriteria

# Mengimport libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Penilaian dari 3 Expert Terhadap Kriteria
pd.read_excel(data_excel, sheet_name= 'expert_kriteria')
```

```python
#STEP 3.3 Tabel Matriks Perbandingan Berpasangan Kriteria
# Menampilkan tabel matriks perbandingan berpasangan berdasarkan penilaian expert
criteria_comparison = pd.read_excel(data_excel, sheet_name= 'kriteria')
criteria_comparison
```

## Step 4 : Eigen Vektor Normalisasi Kriteria

```python
#STEP 4.1 Normalisasi Data Matriks Kriteria

normalization_criteria = criteria_comparison.copy()
for col in ['Jumlah Pelanggan','Jumlah Alat Masuk','Harga Per Alat','Waktu Penyelesaian','Ulasan Pelanggan']:
    normalization_criteria[col] = normalization_criteria[col] / normalization_criteria[col].sum()
normalization_criteria
```

```python
#STEP 4.2 Mencari Bobot Setiap Kriteria

# Menghitung bobot untuk setiap kriteria
Bobot_Jumlah_Pelanggan = (normalization_criteria['Jumlah Pelanggan'][0] + normalization_criteria['Jumlah Alat Masuk'][0] + normalization_criteria['Harga Per Alat'][0] + normalization_criteria['Waktu Penyelesaian'][0] + normalization_criteria['Ulasan Pelanggan'][0])/normalization_criteria['Kriteria'].count()
Bobot_Jumlah_Alat_Masuk = (normalization_criteria['Jumlah Pelanggan'][1] + normalization_criteria['Jumlah Alat Masuk'][1] + normalization_criteria['Harga Per Alat'][1] + normalization_criteria['Waktu Penyelesaian'][1] + normalization_criteria['Ulasan Pelanggan'][1])/normalization_criteria['Kriteria'].count()
Bobot_Harga_Per_Alat = (normalization_criteria['Jumlah Pelanggan'][2] + normalization_criteria['Jumlah Alat Masuk'][2] + normalization_criteria['Harga Per Alat'][2] + normalization_criteria['Waktu Penyelesaian'][2] + normalization_criteria['Ulasan Pelanggan'][2])/normalization_criteria['Kriteria'].count()
Bobot_Waktu_Penyelesaian = (normalization_criteria['Jumlah Pelanggan'][3] + normalization_criteria['Jumlah Alat Masuk'][3] + normalization_criteria['Harga Per Alat'][3] + normalization_criteria['Waktu Penyelesaian'][3] + normalization_criteria['Ulasan Pelanggan'][3])/normalization_criteria['Kriteria'].count()
Bobot_Ulasan_Pelanggan = (normalization_criteria['Jumlah Pelanggan'][4] + normalization_criteria['Jumlah Alat Masuk'][4] + normalization_criteria['Harga Per Alat'][4] + normalization_criteria['Waktu Penyelesaian'][4] + normalization_criteria['Ulasan Pelanggan'][4])/normalization_criteria['Kriteria'].count()

matriks_bobot_kriteria = {'Kriteria' : ['Jumlah Pelanggan','Jumlah Alat Masuk','Harga Per Alat','Waktu Penyelesaian',
                                        'Ulasan Pelanggan'],
                          'Bobot' : [Bobot_Jumlah_Pelanggan, Bobot_Jumlah_Alat_Masuk, Bobot_Harga_Per_Alat,
                                     Bobot_Waktu_Penyelesaian, Bobot_Ulasan_Pelanggan]}
# dataframe bobot kriteria
bobot_kriteria = pd.DataFrame(matriks_bobot_kriteria).set_index('Kriteria')
bobot_kriteria
```

```python
#STEP 4.3 Grafik Bobot Kriteria

kriteria = ['Jumlah Pelanggan','Jumlah Alat Masuk','Harga Per Alat','Waktu Penyelesaian','Ulasan Pelanggan']
nilai_kriteria = [Bobot_Jumlah_Pelanggan, Bobot_Jumlah_Alat_Masuk, Bobot_Harga_Per_Alat, Bobot_Waktu_Penyelesaian, Bobot_Ulasan_Pelanggan]
plt.figure(figsize=(10,4))
plt.bar(kriteria, nilai_kriteria, color = ['blue', 'Green', 'red', 'orange','yellow'])
plt.title('Perbandingan Kriteria')
plt.xlabel('Kriteria')
plt.ylabel('Bobot Kriteria')

plt.show()
```

```python
#STEP 4.4 Hasil Data Matriks

#Hasil akhir dari proses pencarian bobot kriteria adalah :

#Kriteria Jumlah_Alat_Masuk merupakan kriteria yang paling penting dari semua kriteria, dengan bobot 0.472
#1. Kriteria Waktu_Penyelesaian dengan bobot 0.246
#2. Kriteria Harga_Per_Alat dengan bobot 0.116
#3. Kriteria Ulasan_Pelanggan dengan bobot 0.11
#4. Kriteria Jumlah_Pelanggan dengan bobot 0.055
```

```python
#pembagian jumlah perbaris dengan bobot setiap sub kriteria
Hasil_ulasan_pelanggan = jumlah_perbaris_UP.copy()
for x in ['Jumlah Perbaris']:
    Hasil_ulasan_pelanggan[x] = Hasil_ulasan_pelanggan[x] / bobot_ulasan_pelanggan['Bobot']
Emaks_ulasan_pelanggan = Hasil_ulasan_pelanggan[x].sum() / ulasan_pelanggan_comparison['SK Ulasan Pelanggan'].count()
Emaks_ulasan_pelanggan
```

```python
#STEP 6.6.2 Indeks konsistensi sub kriteria Ulasan_Pelanggan

CI_ulasan_pelanggan = (Emaks_ulasan_pelanggan - ulasan_pelanggan_comparison['SK Ulasan Pelanggan'].count()) / (ulasan_pelanggan_comparison['SK Ulasan Pelanggan'].count()-1)
CI_ulasan_pelanggan
```

```python
#STEP 6.6.3 Rasio konsistensi sub kriteria Ulasan_Pelanggan

IR_ulasan_pelanggan = IR.get(ulasan_pelanggan_comparison['SK Ulasan Pelanggan'].count(), None)
if IR_ulasan_pelanggan is not None:
    IR_ulasan_pelanggan
else:
    0
CR_ulasan_pelanggan = CI_ulasan_pelanggan / IR_ulasan_pelanggan
CR_ulasan_pelanggan
```

```python
#STEP 6.6.4 Memeriksa konsistensi data sub kriteria Ulasan_Pelanggan

if CR_ulasan_pelanggan < 0.1:
    print("Data sudah konsisten")
else:
    print("Data belum konsisten, lakukan penilaian ulang")
```

## Step 7 : Perankingan

```python
#STEP 7.1 Rekap Bobot Kriteria dan Sub Kriteria

#Bobot setiap kriteria
bobot_JP = bobot_kriteria.iloc[0, 0]
bobot_JAM = bobot_kriteria.iloc[1, 0]
bobot_HPA = bobot_kriteria.iloc[2, 0]
bobot_WP = bobot_kriteria.iloc[3, 0]
bobot_UP = bobot_kriteria.iloc[4, 0]

#Bobot Sub Kriteria Jumlah_Pelanggan
bobot_dibawah_3 = bobot_jumlah_pelanggan.iloc[0, 0]
bobot_antara_3_5 = bobot_jumlah_pelanggan.iloc[1, 0]
bobot_diatas_5 = bobot_jumlah_pelanggan.iloc[2, 0]

#Bobot Sub Kriteria Jumlah_Alat_Masuk
bobot_dibawah_50 = bobot_jumlah_alat_masuk.iloc[0, 0]
bobot_antara_50_100 = bobot_jumlah_alat_masuk.iloc[1, 0]
bobot_diatas_100 = bobot_jumlah_alat_masuk.iloc[2, 0]

#Bobot Sub Kriteria Harga_Per_Alat
bobot_dibawah_200000 = bobot_harga_per_alat.iloc[0, 0]
bobot_antara_200000_350000 = bobot_harga_per_alat.iloc[1, 0]
bobot_diatas_350000 = bobot_harga_per_alat.iloc[2, 0]

#Bobot Sub Kriteria Waktu_Penyelesaian
bobot_terlambat = bobot_waktu_penyelesaian.iloc[0, 0]
bobot_tepat_waktu = bobot_waktu_penyelesaian.iloc[1, 0]
bobot_sangat_cepat = bobot_waktu_penyelesaian.iloc[2, 0]

#Bobot Sub Kriteria Ulasan_Pelanggan
bobot_kurang_puas = bobot_ulasan_pelanggan.iloc[0, 0]
bobot_puas = bobot_ulasan_pelanggan.iloc[1, 0]
bobot_sangat_puas = bobot_ulasan_pelanggan.iloc[2, 0]
```

```python
#STEP 7.2 Input data 1 periode

#Input Data
input_data = pd.read_excel(data_excel, sheet_name= 'input_data')
input_data
```

```python
#STEP 7.3 Hitung bobot alternatif

bobot_alat_per_kriteria = input_data.copy()

for kriteria in ['Jumlah Pelanggan', 'Jumlah Alat Masuk', 'Harga Per Alat', 'Waktu Penyelesaian', 'Ulasan Pelanggan']:
    if kriteria == 'Jumlah Pelanggan':
        bobot_alat_per_kriteria['Jumlah Pelanggan'] = np.where(bobot_alat_per_kriteria[kriteria] < 3, bobot_JP * bobot_dibawah_3,
                                                         np.where(bobot_alat_per_kriteria[kriteria] > 5, bobot_JP * bobot_diatas_5,
                                                         bobot_JP * bobot_antara_3_5))
    elif kriteria == 'Jumlah Alat Masuk':
        bobot_alat_per_kriteria['Jumlah Alat Masuk'] = np.where(bobot_alat_per_kriteria[kriteria] < 50, bobot_JAM * bobot_dibawah_50,
                                                         np.where(bobot_alat_per_kriteria[kriteria] > 100, bobot_JAM * bobot_diatas_100,
                                                         bobot_JAM * bobot_antara_50_100))
    elif kriteria == 'Harga Per Alat':
        bobot_alat_per_kriteria['Harga Per Alat'] = np.where(bobot_alat_per_kriteria[kriteria] < 200000, bobot_HPA * bobot_dibawah_200000,
                                                         np.where(bobot_alat_per_kriteria[kriteria] > 350000, bobot_HPA * bobot_diatas_350000,
                                                         bobot_HPA * bobot_antara_200000_350000))
    elif kriteria == 'Waktu Penyelesaian':
        bobot_alat_per_kriteria['Waktu Penyelesaian'] = np.where(bobot_alat_per_kriteria[kriteria] == 'Terlambat', bobot_WP * bobot_terlambat,
                                                         np.where(bobot_alat_per_kriteria[kriteria] == 'Tepat Waktu', bobot_WP * bobot_tepat_waktu,
                                                         bobot_WP * bobot_sangat_cepat))
    elif kriteria == 'Ulasan Pelanggan':
        bobot_alat_per_kriteria['Ulasan Pelanggan'] = np.where(bobot_alat_per_kriteria[kriteria] == 'Kurang Puas', bobot_UP * bobot_kurang_puas,
                                                         np.where(bobot_alat_per_kriteria[kriteria] == 'Puas', bobot_UP * bobot_puas,
                                                         bobot_UP * bobot_sangat_puas))
bobot_alat_per_kriteria
```

```python
Bobot_Plug = (bobot_alat_per_kriteria['Jumlah Pelanggan'][0] + bobot_alat_per_kriteria['Jumlah Alat Masuk'][0] + bobot_alat_per_kriteria['Harga Per Alat'][0] + bobot_alat_per_kriteria['Waktu Penyelesaian'][0] + bobot_alat_per_kriteria['Ulasan Pelanggan'][0])
Bobot_Thread_Plug = (bobot_alat_per_kriteria['Jumlah Pelanggan'][1] + bobot_alat_per_kriteria['Jumlah Alat Masuk'][1] + bobot_alat_per_kriteria['Harga Per Alat'][1] + bobot_alat_per_kriteria['Waktu Penyelesaian'][1] + bobot_alat_per_kriteria['Ulasan Pelanggan'][1])
Bobot_Thread_Ring = (bobot_alat_per_kriteria['Jumlah Pelanggan'][2] + bobot_alat_per_kriteria['Jumlah Alat Masuk'][2] + bobot_alat_per_kriteria['Harga Per Alat'][2] + bobot_alat_per_kriteria['Waktu Penyelesaian'][2] + bobot_alat_per_kriteria['Ulasan Pelanggan'][2])
Bobot_Instrument = (bobot_alat_per_kriteria['Jumlah Pelanggan'][3] + bobot_alat_per_kriteria['Jumlah Alat Masuk'][3] + bobot_alat_per_kriteria['Harga Per Alat'][3] + bobot_alat_per_kriteria['Waktu Penyelesaian'][3] + bobot_alat_per_kriteria['Ulasan Pelanggan'][3])
Bobot_Perankingan = {'Alat' : ['Plug Gauge','Thread Plug Gauge','Thread Ring Gauge','Instrument'],
                     'Bobot' : [Bobot_Plug, Bobot_Thread_Plug, Bobot_Thread_Ring, Bobot_Instrument]}
Perankingan = pd.DataFrame(Bobot_Perankingan)
Perankingan
```

```python
#STEP 7.4 Perankingan alternatif berdasarkan urutan besar masing - masing bobot.

# Menyaring DataFrame berdasarkan kolom 'A' dari yang besar ke yang kecil
Ranking = Perankingan.sort_values(by='Bobot', ascending=False)
# Menyusun ulang indeks dari 1 hingga terakhir
Ranking.reset_index(drop=True, inplace=True)
# Menambahkan kolom baru 'peringkat' berdasarkan nilai kolom 'A'
Ranking['Peringkat'] = Ranking['Bobot'].rank(ascending=False, method='min')
Ranking
```

```python
#STEP 7.5 Grafik Bobot perankingan alternatif

plt.figure(figsize=(8,4))
plt.bar('Alat', 'Bobot', color = ['dimgray','navy','purple','maroon'], data=Ranking)
plt.title('Perbandingan Alternatif Setiap Alat')
plt.xlabel('Alat')
plt.ylabel('Bobot Alternatif')
plt.show()
```
